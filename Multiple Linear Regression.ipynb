Multiple Linear Regression


```python
from warnings import filterwarnings
filterwarnings('ignore')
```

Read the dataset

```python
import pandas as pd
df=pd.read_csv('50_Startups.csv')
df.head()
```

Perform basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

PROFIT is DEPENDENT FEATURE
ADMIN,MKT,RND ARE INDEPENDENT FEATURES
PROFIT ~ ADMIN,MKT,RND
PROFIT_PRED=B0 + B1xADMIN + B2xRND + B3xMKT

SEPARATE X AND Y

```python
X=df[['RND','ADMIN','MKT']]
Y=df[['PROFIT']]
```

```python
X.head()
```

```python
Y.head()
```

Creating a Linear Regression model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,Y)
```

check r2 score for the model

```python
model.score(X,Y)
```

Get coefficents and intercept for the above model

```python
model.coef_
```

```python
model.intercept_
```

PROFIT_PRED=50122.19298987 + 805.7XB1 - 26.8XB2 + 27.2XB3

3 conclusions:
1. if RND increases by 1000,then PROFIT increases by 805.7
2. if ADMIN increases by 1000,then PROFIT decreases by 26.8
3. if MKT increases by 1000,then PROFIT increases by 27.2

More the budget invested in RND , higher the impact on profit

Predict the estimated results for X by the model

```python
ypred=model.predict(X)
ypred[0:5]
```

```python
Y.head()
```

Evalualting the model

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
mse=mean_squared_error(Y,ypred)
mse
```

```python
rmse=mse**(1/2)
rmse
```

```python
mae=mean_absolute_error(Y,ypred)
mae
```

```python
mape=mean_absolute_percentage_error(Y,ypred)
mape
```

```python
r2_score(Y,ypred)
```

Above results i get r2 score of 0.9507 which is very good >0.8
