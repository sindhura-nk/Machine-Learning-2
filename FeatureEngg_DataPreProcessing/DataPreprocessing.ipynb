### Create a replacer function to replace missing values

```python
def replacer(r):
    # Get categorical and continuous features
    cat=list(r.columns[r.dtypes=='object'])
    con=list(r.columns[r.dtypes!='object'])
    # Apply forloop on all columns
    for i in r.columns:
        if i in cat:
            md=r[i].mode()[0]
            r[i]=r[i].fillna(md)
        else:
            mn=r[i].mean()
            r[i]=r[i].fillna(mn)
    print('Missing values replaced in dataframe')
```
