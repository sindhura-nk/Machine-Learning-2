## Separate X and Y .
## X ~ independent features, remove statistically insignificant columns, remove target feature
## Y ~ Dependent feature .. Target Feature

```python
X = df.drop(columns=['id','Weight'])
Y = df[['Weight']]
```

```python
X.head()
```

```python
Y.head()
```

## Categorical and continuous features for X

```python
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
con
```

```python
cat
```

## Data Preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent')),
                           ('ordinal',OrdinalEncoder())])
```

```python
pre = ColumnTransformer([('num',num_pipe,con),
                         ('cat',cat_pipe,cat)]).set_output(transform='pandas')
```

```python
pre
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

## Feature selection using Backward Elimination

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
lr = LinearRegression()
sel = SequentialFeatureSelector(lr,direction='backward',n_features_to_select='auto')
sel.fit_transform(X_pre,Y)
sel_cols = sel.get_feature_names_out()
```

```python
sel_cols
```

```python
sel_cols[0]
```

```python
sel_cols[0].split('__')
```

```python
sel_cols[0].split('__')[1]
```

```python
imp_cols= []
for i in sel_cols:
    col= i.split('__')[1]
    imp_cols.append(col)
imp_cols
```

```python
X_sel = X[imp_cols]
X_sel
```

### separate cat and con features

```python
cat_sel = list(X_sel.columns[X_sel.dtypes=='object'])
con_sel = list(X_sel.columns[X_sel.dtypes!='object'])
```

```python
cat_sel
```

```python
con_sel
```

## create a final pipeline

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent')),
                            ('ohe',OneHotEncoder(handle_unknown='ignore',sparse_output=False))])
```

```python
pre2 = ColumnTransformer([('num',num_pipe2,con_sel),
                          ('cat',cat_pipe2,cat_sel)]).set_output(transform='pandas')
```

```python
pre2
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre.head()
```

## split the data into training and testing

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_sel_pre,Y,test_size=0.2,random_state=21)
```

```python
xtrain.head()
```

```python
xtest.head()
```

## Build Linear Regression model

```python
model = LinearRegression()
model.fit(xtrain,ytrain)
```

## Evaluate model in train and test

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

## Backward elimination has lesser r2 score in test when compared to forward selection
## Getting Adjusted R2 for train data


```python
xtrain.shape
```

```python
xtrain.shape[0]
```

```python
xtrain.shape[1]
```

```python

def adj_r2(model, xtrain, ytrain):
    # Get r2 score
    r2 = model.score(xtrain, ytrain)
    # Get Number of Rows
    N = xtrain.shape[0]
    # Get number of columns
    p = xtrain.shape[1]
    # Get adjr2 
    num = (1-r2)*(N-1)
    den = N-p-1
    r2a = 1 - num/den
    return r2a 
```

```python

adj_r2(model, xtrain, ytrain)
```

## Function to get mse, rmse, mae, r2

```python
def evaluate_model(model, x, y):
    from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
    # Predict results for train and test
    ypred = model.predict(x)
    # Calulate the metrics
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    r2 = r2_score(y, ypred)
    # Print above results
    print(f'Mean Squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute Error : {mae:.2f}')
    print(f'R2 Score : {r2:.4f}')
```

```python
evaluate_model(model, xtrain, ytrain)
```
