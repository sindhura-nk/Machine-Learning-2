```python
from warnings import filterwarnings
filterwarnings('ignore')
```

## Step 1 :Data ingestion (REading the data)

```python
import pandas as pd
df=pd.read_csv('Cars93.csv',na_values=['','NA'],keep_default_na=False)
df.head()
```

```python
n=df.isna().sum()
n[n>0]
```

### Separate target feature and independent feature

```python
X=df.drop(columns=['id','Weight'])
Y=df[['Weight']]
```

```python
X.head()
```

```python
Y.head()
```

## Perform preprocessing on X

```python
# Separate cat and con
cat=list[X.columns[X.dtypes =='object']]
con=list(X.columns[X.dtypes !='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline #Automate the steps
from sklearn.impute import SimpleImputer # Perform missing value treatment. it replaces the data
from sklearn.preprocessing import OneHotEncoder,StandardScaler # preprocessing the data
from sklearn.compose import ColumnTransformer # combine both cat and con features
```

```python
num_pipe = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),('scaler',StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent')),('ohe',OneHotEncoder(handle_unknown='ignore',sparse_output=False))])
```

```python
pre=ColumnTransformer([('num',num_pipe,con),
                       ('cat',cat_pipe,cat)]).set_output(transform='pandas')
```

```python
pre
```

```python
X_pre=pre.fit_transform(X)
```

```python
X_pre.head()
```

Save preprocessed data to csv

```python
X_pre.to_csv('Preprocessed.csv',index=False)
```

Perform train test split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_pre,Y,test_size=0.2,random_state=21)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

```python
xtest.shape
```

## Step 6 :Model Building

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
```

## Step 7: Evaluate the model

    1.MSE
    2.RMSE
    3.MAE
    4.R2 score

```python
def evaluate_model(xtrain,ytrain,xtest,ytest,model):
    # r2 scores
    r2_train=model.score(xtrain,ytrain)
    r2_test=model.score(xtest,ytest)
    # Get prediction on test and train
    ypred_train = model.predict(xtrain)
    ypred_test = model.predict(xtest)
    # Getting mse,rmse,mae
    from sklearn.metrics import mean_absolute_error,mean_squared_error
    mse_train = mean_squared_error(ytrain,ypred_train)
    mse_test = mean_squared_error(ytest,ypred_test)
    rmse_train = mse_train**(1/2)
    rmse_test = mse_test**(1/2)
    mae_train = mean_absolute_error(ytrain,ypred_train)
    mae_test = mean_absolute_error(ytest,ypred_test)
    # print above results
    print("Training Evaluation Results \n")
    print(f'MSE : {mse_train:.2f}')
    print(f'RMSE : {rmse_train:.2f}')
    print(f'MAE : {mae_train:.2f}')
    print(f'R2 : {r2_train:.2f}')
    print('\n ========================================= \n')
    print("Testing Evaluation Results \n")
    print(f'MSE : {mse_test:.2f}')
    print(f'RMSE : {rmse_test:.2f}')
    print(f'MAE : {mae_test:.2f}')
    print(f'R2 : {r2_test:.4f}')    
```

```python
evaluate_model(xtrain,ytrain,xtest,ytest,model)
```

## Because r2 score in testing is also more than 0.8 above model is a good model
